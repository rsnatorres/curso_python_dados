### 1) .loc[]

The simples way. Ideal when we have few options

```python
df1.loc[ (df1['idade'] < 20) , 'idade'] # testing some condition
# note that .loc[] retrieves two informations: the index and the values
```

```python
number_1 = 3 # test with 10
number_2 = 9

if number_1 < number_2:
  print('number_1 is the smallest')
else: # if 'if statement' is false, what should the programm do
  print('number_2 is the smallest ')
```

```python
# What happens if the two numbers provided are the same?
number_1 = 9
number_2 = 9

# if statements only proceeds with chainned command if sentence provided is True
if number_1 < number_2:
  print('number_1 is the smallest')
else:
  print('number_2 is the smallest ')
```

```python
# all possible conditions
number_1 = 3
number_2 = 9

if number_1 < number_2:
  print('number_1 is the smallest')
elif number_1 == number_2: # elif is an EXCLUSIVE option: can only be executed if none of the previouslly 'if statements' were true
  print('both numbers are the same')
else:
  print('number_2 is the smallest ')
```

## 3) np.select()

3 steps:


*   1) set the conditions
*   2) set the values you want
*   3) apply func

The most efficient solution (memory optimization)

```python
conditions = [ df1['idade'] < 20,
              (df1['idade'] >= 20) & (df1['idade'] < 40),
              (df1['idade'] >= 40) & (df1['idade'] < 60),
               df1['idade'] > 60]
```

```python
values = ['ate_20',
          'entre_20_40',
          'entre_40_60',
          'mais_que_60']
```

```python
np.select(conditions, values)
```

```python
df1['grupo_idade3'] = np.select(conditions, values)
df1['grupo_idade3']
```

## 3.1 Difference between if and elif

```python
# In previouslly examples we did:
number_1 = 3
number_2 = 9

if number_1 < number_2: # if condition 1 is met, the program will stop there, and do what command inside it requires
  print('number_1 is the smallest')
elif number_1 == number_2:
  print('both numbers are the same')
else:
  print('number_2 is the smallest ')
```

```python
# if we use ifs instead of elifs --> program will continue and check every if
number_1 = 3
number_2 = 9

if number_1 < number_2:
  print('number_1 is the smallest')
if number_1 + 2 < number_2:
  print('number_1 plus 2 is still the smallest')
if number_1 + 5 < number_2:
  print('number_1 is the smalles event if we add 5')
```

```python
# if elif is used, no matter if condition setted is true,
# if some condition before it is true, program will stop check conditions
number_1 = 3
number_2 = 9

if number_1 < number_2:
  print('number_1 is the smallest')
elif number_1 + 2 < number_2:
  print('number_1 plus 2 is still the smallest')
elif number_1 + 5 < number_2:
  print('number_1 is the smalles event if we add 5')
```
