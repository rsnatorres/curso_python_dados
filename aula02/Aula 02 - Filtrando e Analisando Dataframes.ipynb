### 3.1.3 .pivot_table()

```python
# whats the average wage for each occupation?
df1.pivot_table(index = 'descricao_ocupacao', # var to group --> some string or categorical var --> # ANALYSIS UNIT
               values = 'salario_mensal', # var to be calculated --> some continuous var
               aggfunc = 'mean' # the function to apply
               )
```

```python
# applying chainned commands
df1.pivot_table(index = 'descricao_ocupacao', # var to group
               values = 'salario_mensal', # var to be calculated
               aggfunc = 'mean' # the function to apply
               ).sort_values(by='salario_mensal').round(2)
# obs: the chainned method is being applied into the pivot_table, not the dataframe
```

```python
# we can assign results into a new dataframe
tab_salarios = df1.pivot_table(index = 'descricao_ocupacao', # var to group
                              values = 'salario_mensal', # var to be calculated
                              aggfunc = 'mean' # the function to apply
                              )
tab_salarios.head(3)
# BE AWARE --> the index var is the index of the new df, it is not a column
```

Have in mind:

Index --> Categorical var

Values --> Continous var

Aggfunc --> the function ('mean', 'count', 'median', 'std', 'sum')

Testing different Indexes, values and aggfunc

#### Adding another aggfunctions

We could also add a measure of dispersion of data

```python
# the new aggfunctions will be applied for the values specified inside pivot_table
df1.pivot_table(index = 'descricao_ocupacao',
                values = 'salario_mensal',
                aggfunc = 'count'
                )
# count --> does counting in respect to the column defined in 'values'
# equivalent --> we have X salaries informed for profession Y
```

```python
# adding more than one function simultaneously
tab_salarios = df1.pivot_table(index = 'descricao_ocupacao',
                              values = 'salario_mensal',
                              aggfunc = ['mean', 'std', 'median', 'count', 'min', 'max', 'sum']
                              )
tab_salarios.head(3)
```

#### Adding vars to calculation

```python
df1.pivot_table(index = 'descricao_ocupacao', # var to group
                values = ['salario_mensal', 'idade'], # var to be calculated
                aggfunc = 'mean' # the function to apply
                ).round(2)
```

```python
df1.pivot_table(index = 'descricao_ocupacao', # var to group
                values = ['salario_mensal', 'idade'], # var to be calculated
                aggfunc = 'mean' # the function to apply
                ).sort_values(by='idade').round(2)
```

#### Adding levels to index

Building a multi-level index dataframe

```python
# Index level is set by order of values
df1.pivot_table(index = ['descricao_grau_instrucao', 'sexo'], # var to group
                values = 'salario_mensal', # var to be calculated
                aggfunc = 'mean' # the function to apply
                ).round(2)
```

#### Adding Columns

```python
# generating one column for each value in columns
df1.pivot_table(index = 'descricao_grau_instrucao', # var to group
                values = 'salario_mensal',
                columns='sexo', # breaks index in two groups to be displayed as different columns
                aggfunc = 'mean' # the function to apply
                ).round(2)
```

#### specify aggfunc for values

But how to calculate an aggfun for just a specific var?

passing a dictionary structure inside aggfunc: keys will be the columns and the values will be the desire aggfunc

```python
df1.pivot_table(index = 'cnae_2_secao', # var to group
                values = ['salario_mensal', 'idade'], # var to be calculated
                aggfunc = ['mean', 'sum']).head()
```

```python
# adding dictionay structure into aggfunc argument
# with this its possible to specify the aggfunc for each column
# the structure must meet the pattern {'column1' : 'aggfunc1', 'column2': 'aggfun2'}
tab_salario = df1.pivot_table(index = ['cnae_2_secao'], # var to group
                                    values = ['salario_mensal', 'idade'], # var to be calculated
                                    aggfunc = {'salario_mensal': ['sum', 'mean'],
                                              'idade': 'mean'}
                                    ).round(2)
tab_salario.head()
```

## 5.5 Using percentiles in .pivot_table()

```python
# lets take the below table as example --> suppose we wanted to have percentiles of 'salario_mensal'
df1.pivot_table(index = 'descricao_ocupacao', # var to group
                values = 'salario_mensal', # var to be calculated
                aggfunc = 'median' # the function to apply
                ).round(2)
```

```python
# first we must be sure we imported numpy
import numpy as np
```

```python
# we must build the percentile functions --> yes, what an inconvinient way, but what I can do?
def per_25(df):
        return np.percentile(df, 25)
def per_75(df):
        return np.percentile(df, 75)
```

```python
# and them is just to apply them withou quetoes inside aggfunc along the other aggfuncs we have viewed
df1.pivot_table(index = 'descricao_ocupacao', # var to group
                values = 'salario_mensal', # var to be calculated
                aggfunc = [per_25, 'median', per_75]# the function to apply
                ).round(2)
```
