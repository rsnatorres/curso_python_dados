## 3.Datetime

Pandas has an specific formatt to deal with datetimes in dataframe.

It is called DatetimeIndex. It can be usefull for operations which can be only applied with dates.

```python
# In this dataframe dates are formatted as integers and not datetimeIndex
df1['ano'].dtype, df1['mes'].dtype
```

```python
# first we must pass the columns which has info about dates into strings
df1['ano'] = df1['ano'].astype(str)
df1['mes'] = df1['mes'].astype(str)
```

```python
# them we concatenate them with whatever characters you want to use as separator between month and year
df1['ano_mes'] = df1['ano'] + '/' + df1['mes']
df1['ano_mes']
```

### pd.to_datetime()

This method will try to infer datetime format accordingly with the data that is passed to it.

for formats see python docs:

https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes

```python
# in this case it infered correctely
pd.to_datetime(df1['ano_mes'])
# datetime type will always show some day. Even if the original data had none.
# in this case it will be the first day of the month
```

```python
# it is always more secure to specify format parameter when transforming a column into a datetime format
# this parameter must show how data is formatted in its original form
pd.to_datetime(df1['ano_mes'], format = '%Y/%m')
# %Y --> stands for complete year format (i.e. 2022, 2021, not 22 or 21)
# %m --> stands for months with zero-padded fill (i.e 01 for january, not just 1)
# to check the sintax which format accepts see https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes
```

```python
# format must be specified as origin data format
df1['data']  = pd.to_datetime(df1['ano_mes'], format = '%Y/%m') # input format
```

```python
df1['data'] # check dtype
```
