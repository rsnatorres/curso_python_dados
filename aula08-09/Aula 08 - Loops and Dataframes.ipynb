```python
# note how we are repeating the same command over and over again
df1['descricao_grau_instrucao'] = df1['descricao_grau_instrucao'].str.strip()
df1['descricao_ocupacao'] = df1['descricao_ocupacao'].str.strip()
df1['sigla_uf'] = df1['sigla_uf'].str.strip()
df1['cnae_2_secao'] = df1['cnae_2_secao'].str.strip()
```

Steps for creating a loop:

1. Find the pattern
2. create a variable to replace what has been changing between commands
3. Take what has been changing and create a list with it
4. build the loop

```python
# this is the basic structure of all previous transformations
#df1[label_coluna] = df1[label_coluna].str.strip() # always choose some intuitive name for the object to be replaceble
```

```python
# this is necessary, but somehow hard to check
for label_coluna in ['descricao_grau_instrucao', 'descricao_ocupacao', 'sigla_uf', 'cnae_2_secao']:
  df1[label_coluna] = df1[label_coluna].str.strip()
```

```python
# this command is better for checking
for label_coluna in ['descricao_grau_instrucao', 'descricao_ocupacao', 'sigla_uf', 'cnae_2_secao']:
  df1[label_coluna] = df1[label_coluna].str.upper()
```

# 4.Wide and Long dataframes

```python
tab_salarios = df1.pivot_table(index = ['cnae_2_secao', 'mes'],
                              values = ['saldo_movimentacao', 'salario_mensal'],
                              aggfunc = {'saldo_movimentacao': 'count',
                                        'salario_mensal': ['mean', 'median']}
                              )
```

```python
tab_salarios.reset_index(inplace=True)
```

```python
tab_salarios.head()
```

```python
tab_salarios.columns
```

```python
tab_salarios.columns = ['cnae_2_secao', 'mes', 'Média Salarial', 'Mediana Salarial', 'Nº de Contratações']
```

```python
tab_salarios.head(10)
```

## 4.1 pd.melt()

```python
pd.melt(tab_salarios,
        id_vars = ['cnae_2_secao', 'mes'], # Identifiers variables --> varibles used to segment our data
        value_vars = ['Média Salarial', 'Mediana Salarial', 'Nº de Contratações'] # columns which contains data to be ploted
        )
```

```python
# if we moved cnae_2_secao to value_vars --> we would end up with a mixed tipe values in value column
pd.melt(tab_salarios,
        id_vars = ['mes'],
        value_vars = ['Média Salarial', 'Mediana Salarial', 'Nº de Contratações', 'cnae_2_secao']
        )
```

```python
tab_salarios_long = pd.melt(tab_salarios,
                            id_vars = ['cnae_2_secao', 'mes'], # varibles used to segment our data
                            value_vars = ['Média Salarial', 'Mediana Salarial', 'Nº de Contratações'] # columns which contains data to be ploted
                            )
```

```python
tab_salarios_long
```

```python
fig = px.line( tab_salarios_long.query("cnae_2_secao == 'A' & variable == 'Nº de Contratações' "),
              x = 'mes',
              y = 'value')

fig.write_image("Nº de Contratações no Setor A em 2020.png")

fig
```

```python
tab_salarios_long['cnae_2_secao'].unique()
```

```python
# with this code we generated figures for all sectors, but then we had only one variable: 'Nº de Contratações'
for setor in tab_salarios_long['cnae_2_secao'].unique():
  fig = px.line( tab_salarios_long.query("cnae_2_secao == 'A' & variable == 'Nº de Contratações' "),
                x = 'mes',
                y = 'value')

  fig.write_image(f"Nº de Contratações no Setor {setor} em 2020.png")
```

```python
tab_salarios_long.head(3)
```

```python
# but now, we have 3 variables to plot for each sector
tab_salarios_long['variable'].unique()
```

## 4.2 nested loops

```python
# a nested loop
for variavel in tab_salarios_long['variable'].unique():
  for setor in tab_salarios_long['cnae_2_secao'].unique():

    fig = px.line( tab_salarios_long.query("variable == @variavel & cnae_2_secao == @setor"),
                  x = 'mes',
                  y = 'value')

    fig.write_image(f"/content/drive/My Drive/cursos/disciplina_python_dados/images/{variavel} no Setor {setor} em 2020.png")
```
