# 1.Introduction

**Working with Google Colab**

Google Colaboratory (or just 'Colab' for close friends) is an IDLE for python. IDLE stands for Integrated Development and Learning Environment (IDLE). It is just a graphic interface to make your life easier in lots of ways when programming.

Colab was inspired by Jupyter Notebook Project, which mantains the most famous IDLE for Python. The main difference between the two is that Jupyter Notebook works directly with your computer's hardware (memmory and processing) and Colab works with Google's Data Centers hardware. So one runs in your machine, the other runs on cloud.

Pros and Cons of COLAB:
Pros:
1.   Do not need to install python on your computer, neither any of its libraries
2.   Skip of most packages version problems
3.   Code is automatically saved (But ALWAYS check)

Cons:
1.   Colab has limited memmory allocation. If you surpass it, Colab will restar the code and you can lose variables
2.   Colab has limited time for continuously run code (12 hours)


Shorcuts of Colab:

1. command? --> open documentation
2. comand?? --> open source code
3. Alt + Enter --> run command in selected cell

## 4.3 Categorical

In this dataframe, the only numeric columns are:

*   idade
*   salario_mensal
*   horas_contratuais


All other non-string variables, besides being storaged as numbers, are categorical vars: it does not make sense to do arithmetic operations with those vars, because they represent groups and not continuously values

````python
df.head(3)
````

````python
df['grau_instrucao']
# grau_instrução is not a numeric information, but a categorical information
````

#### 4.3.1 .astype('category')

````python
df['grau_instrucao'].astype('category')
````

````python
df['grau_instrucao'] = df['grau_instrucao'].astype('category')
````

````python
df.info() # cheking new type
````

#### 4.3.2 pd.Categorical()

This is a more complete method/function for categorical transformation than .astype('category'). With this function you can define categories order

````python
# come back to original type to transform it again
df['grau_instrucao'] = df['grau_instrucao'].astype(int)
````

````python
df['grau_instrucao'].dtype
````

````python
pd.Categorical(df['grau_instrucao'])
````

````python
# 80 stands for especialization or mba, it be before a master's degree (number 10)
# Verify this proposition of order in 'layout_caged.xlsx'
df['grau_instrucao'] = pd.Categorical(df['grau_instrucao'],
                                    categories = [1, 2, 3, 4, 5, 6, 7, 8, 9, 80, 10, 11, 99],
                                    ordered=True)
````

````python
# see the new information at the end of output, it shows the categories order
df['grau_instrucao']
````

````python
# to get an ordered list with categories
df['grau_instrucao'].cat.categories
````

## 5.3 Pandas Options

We have a number of pandas options in order to custom the way it works

There are two ways to set pandas options, we'll see them both

````python
# setting max numbers of columns (default is 10)
pd.set_option('display.max_columns', 20) # way 1
pd.options.display.max_columns = 20 # way 2

# setting max numbers of rows (default is 10)
pd.set_option('display.max_rows', 20) # way 1
pd.options.display.max_rows = 20 # way 2

# setting max lenght of characters in a column (default is 50)
pd.set_option('display.max_colwidth', 100) # way 1
pd.options.display.max_colwidth = 100 # way 2

# setting max decimal digits for floats (default is 6)
# obs: this costumization is just for display, it does not alter the original data
pd.set_option('display.precision', 4) # way 1
pd.options.display.precision = 4 # way 2
````

## 5.3 Other read_csv() args

Reading csv files can be an easy task, but sometimes it can be a harsh one.

The following command show some commonly used arguments when just read_csv(file.csv) can't do the job

````python
pd.read_csv('/content/drive/My Drive/cursos/disciplina_python_dados/caged_mg_2020_filtrado.csv',
            sep = ',', header = 0,
            decimal = '.', # use ',' for european (except uk) or brasizilian data
            encoding='UTF-8', # In rarely cases 'Latin-1' work better for portuguese strings
            usecols = ['cnae_2_secao', 'salario_mensal'], # to import just the specified columns
            dtype = {'cnae_2_secao': str, # set data type previouslly
                     'salario_mensal': float},
            na_values = ['null', 'nula', 'nulo'] # consider all this values as null
    )
````
